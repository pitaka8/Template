# 위젯 설정

```python
from tkinter import *
from tkinter import messagebox as msg

# window 설정
window=Tk()
window.title("Tkinter 연습")
window.geometry('400x300')

def myFunc():
    pass

def myFunc1():
    msg.showinfo("입력버튼", "값을 입력해주세요")
    
def myFunc2():
    if chk.get()==0:
        msg.showinfo("","선택해제")
    else:
        msg.showinfo("","선택")
        
def myFunc3():
    if varRb.get()==1:
        msg.showinfo("","1번선택")
    elif varRb.get()==2:
        msg.showinfo("","2번선택")
    else:
        msg.showinfo("","3번선택")
        
#Label 입력
label1=Label(window, text="Label 연습", font=("고딕체",20),fg='black', bg="magenta",width=10,height=2,
               anchor='center')

# Button 입력
btn1=Button(window,text="입력버튼1",command=myFunc)
btn2=Button(window,text="입력버튼2",command=myFunc)
btn3=Button(window,text="입력버튼3",command=myFunc)



#Check Button 입력
chk=IntVar()
cb1=Checkbutton(window,text="옵션을 선택", variable=chk,command=myFunc2)

#RadioButton 입력
varRb=IntVar()
rb1=Radiobutton(window,text="1번",variable=varRb,value=1,command=myFunc3)
rb2=Radiobutton(window,text="2번",variable=varRb,value=2,command=myFunc3)
rb3=Radiobutton(window,text="3번",variable=varRb,value=3,command=myFunc3)


label1.place(x=0,y=0)
btn1.pack(side=TOP,padx=10,pady=10,ipadx=10,ipady=10)
btn2.pack(side=TOP,padx=10,pady=10,ipadx=10,ipady=10)
btn3.pack(side=TOP,padx=10,pady=10,ipadx=10,ipady=10)
cb1.place(x=20,y=70)
rb1.pack()
rb2.pack()
rb3.pack()
window.mainloop()
```

#  메뉴창 만들기

```python
from tkinter import *
from tkinter import messagebox as msg

def func_open():
    pass

def func_exit():
    window.quit()
    window.destroy()

# window 설정
window=tk.Tk()
window.title("Tkinter 연습")
window.geometry('400x300')

#File 메뉴 생성
mainMenu=Menu(window)
window.config(menu=mainMenu)

fileMenu=Menu(mainMenu)
mainMenu.add_cascade(label="파일",menu=fileMenu)
fileMenu.add_command(label="열기",command=func_open)
fileMenu.add_separator()
fileMenu.add_command(label="종료",command=func_exit)

window.mainloop()
                     
```
